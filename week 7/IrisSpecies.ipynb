```python
# Task 2: Basic Data Analysis
import pandas as pd
import numpy as np

df = pd.read_csv('Iris.csv')
#compting basic statistics
data=df.describe()
print(data)
#performing groupings
grouped=df.groupby('Species').mean()
print(grouped)
# identifying partens or insights from analysis
print("\nPatterns/insights")
print("The Iris has got 3 species")
print("Iris-virginica is the largest of the three species")
```

```python
# Task 3: Data Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Iris.csv')
#linechart trend over samples
plt.figure(figsize=(10, 6))
plt.plot(df['SepalLengthCm'], label='Sepal Length', marker='o')
plt.title('Trend of Sepal Length Over Samples')
plt.xlabel('Sample')
plt.ylabel('Sepal Length')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Bar chart showing the comparison of a numerical value across categories
plt.figure(figsize=(10, 6))
sns.barplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Comparison of Sepal Length Across Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.grid(True)
```

```python
#Histogram of a numerical column to understand its distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['SepalLengthCm'], bins=20, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Scatter plot to visualize the relationship between two numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
plt.title('Relationship between Sepal Length and Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(True)
plt.show()
```
